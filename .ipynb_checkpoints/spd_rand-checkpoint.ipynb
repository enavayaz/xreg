```python
#%env JAX_PLATFORM_NAME=cpu
#%env JAX_ENABLE_X64=True
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as lg
import jax
from util_pred import fit_poly_dc, cov_mat
from morphomatics.manifold import SPD, PowerManifold
from morphomatics.stats import ExponentialBarycenter
from timeseries.verification_metrics import errfun
#from timeseries.util_opt import *
from timeseries.model import VelocityEnsemble, Reg, RidgeReg, ARMA, WeightedAverage
from timeseries.mainfull import pred
from matplotlib.patches import Ellipse

def cov_to_ellipse_params(cov):
    """Convert 2x2 covariance matrix to ellipse parameters (a, b, angle)."""
    eigval, eigvec = np.linalg.eigh(cov)  # eigh for symmetric matrices
    angle = np.degrees(np.arctan2(eigvec[1, 1], eigvec[0, 1]))
    return np.sqrt(eigval[0]), np.sqrt(eigval[1]), angle


def vis_ellipse(cov_sequences, centers=None):
    """
        Visualize sequences of 2x2 covariance matrices as concentric ellipses.

        Parameters:
        -----------
        cov_sequences : list of arrays
            List of 1 or 2 sequences. Each sequence is an array of shape (n, 2, 2)
            containing positive definite symmetric matrices.
        centers : array-like, optional
            Centers for ellipses. If None, places them horizontally.

        Returns:
        --------
        fig, ax : matplotlib figure and axis
        """
    # Ensure cov_sequences is a list
    if not isinstance(cov_sequences, list):
        cov_sequences = [cov_sequences]

    n_sequences = len(cov_sequences)
    colors = ['b', 'r']
    linestyles = ['-', '--']

    # Convert all covariance matrices to ellipse parameters
    ellipse_params = []
    for seq in cov_sequences:
        params = np.array([cov_to_ellipse_params(cov) for cov in seq])
        ellipse_params.append(params)

    # Setup centers
    max_radius = max([np.max(params[:, :2]) for params in ellipse_params])
    n_ellipses = len(ellipse_params[0])

    if centers is None:
        spacing = 2.5 * max_radius  # Space ellipses to avoid overlap
        centers = np.array([[(i + 1) * spacing, 0] for i in range(n_ellipses)])
        centers = [centers] * n_sequences
    else:
        centers = [np.asarray(c) for c in centers] if isinstance(centers, list) else [np.asarray(centers)]
        if len(centers) == 1 and n_sequences > 1:
            centers = centers * n_sequences

    # Create plot
    fig, ax = plt.subplots(figsize=(10, 6), subplot_kw={'aspect': 'equal'})

    # Set limits
    all_centers = np.vstack(centers)
    ax.set_xlim(all_centers[:, 0].min() - max_radius,
                all_centers[:, 0].max() + max_radius)
    ax.set_ylim(all_centers[:, 1].min() - max_radius,
                all_centers[:, 1].max() + max_radius)

    # Draw ellipses
    for i, (params, ctr) in enumerate(zip(ellipse_params, centers)):
        for j, (a, b, angle) in enumerate(params):
            ellipse = Ellipse(ctr[j], width=2 * a, height=2 * b, angle=angle,
                              edgecolor=colors[i], linestyle=linestyles[i],
                              fill=False, linewidth=1.5)
            ax.add_patch(ellipse)

    ax.grid(True, alpha=0.3)
    #ax.set_xlabel('x')
    #ax.set_ylabel('y')

    return fig, ax

# Global constants
n_subj, n_points = 20, 30
dim = 2
M = SPD(dim)
dist = jax.jit(M.metric.dist)
#sph_log = jax.jit(M.metric.log)
err = errfun(M.metric.dist)
avg = jax.jit(M.metric.pairmean)

# Random SPD Trajectories
def randA_spd(dim=dim):
    A = np.random.randn(dim, dim)
    return np.dot(A, A.T)

def rand_spd_exp(dim=dim):
    from scipy.linalg import expm
    A = np.random.randn(dim, dim)
    A_sym = (A + A.T) / 2  # make symmetric
    return expm(A_sym)

def rand_spd(dim=dim):
    D = np.eye(dim)
    D[0, 0] = np.random.uniform(.5, 1.5)  # or exponential, etc.
    D[1, 1] = np.random.uniform(2.5, 3.5)
    #Q, _ = np.linalg.qr(np.random.randn(dim, dim))  # random orthogonal
    phi = np.random.uniform(-np.pi/20, np.pi/20)
    Q = np.array([[np.cos(phi), -np.sin(phi)], [np.sin(phi), np.cos(phi)]])
    return Q @ D @ Q.T

def generate_rand_trj(n_subj=n_subj, n_points=n_points):
    Y =[]
    for i in range(n_subj):
        spd = np.zeros((n_points, dim, dim))
        for j in range(n_points):
            spd[j] = rand_spd(dim)
        Y += [spd]
    return Y

def pred_old(Y_test, model, iterative, n_learn=1, n_pred=2):
    Y_pred, n_test = [], len(Y_test)
    MAE, Persist_MAE = [], []
    # len_x = len_mean
    # x = np.linspace(0.0, 1.0, len_x)
    for k in range(n_test):
        y_test = Y_test[k]
        y_pred, mae, persist_mae = [], [], []
        for n in range(n_learn, len(y_test)):
            y_learn = y_test[:n]
            #if n > n_cp: y_learn = y_test[n - n_cp:n]
            len_x = n + n_pred
            x = np.linspace(0.0, 1.0, len_x)
            model = model.fit(x[:len(y_learn)], y_learn)
            #b = model.trend.control_points
            p = np.array(model.predict(x[n:len_x], iterative=iterative))
            #p[-1] = M.metric.geopoint(y_learn[-1], p[-1], .5)
            y_pred.append([p[0]])
            mae.append(dist(p[0], y_test[n]))
            persist_mae.append(dist(y_test[n-1], y_test[n]))
        Y_pred.append(np.array(y_pred))
        MAE.append(mae)
        Persist_MAE.append(persist_mae)
    # Statistics
    aLL_MAE = np.concatenate(MAE)  # cross-trj (pooled)
    # Compute MASE per trajectory (not pooled)
    mase_per_track = [np.mean(MAE[k]) / np.mean(Persist_MAE[k])
                      for k in range(n_test) if np.mean(Persist_MAE[k]) > 0]
    mase_mean = np.mean(mase_per_track)
    mase_std = np.std(mase_per_track)
    #mase_pooled = np.mean(all_mae) / np.mean(all_persist)  # less popular
    skill = 100 * (1 - mase_mean)
    print(f'MAE (forcast-average): {np.mean(aLL_MAE):.4f}')
    print(f'MASE (per-track): {mase_mean:.4f} ± {mase_std:.4f}')
    print(f'Skill:            {skill:.1f}%')
    return Y_pred, MAE

def compute_rmae_mu(M, y_true: np.ndarray, y_pred: np.ndarray, exp_barycenter_fun) -> float:
    """
    Computes rMAE_mu: (Average error) / (Average deviation from the mean).

    Parameters
    ----------
    M : manifold.Manifold
        The manifold object (e.g., SPD(2)).
    Y_true : np.ndarray
        True manifold points (shape: N, ...).
    Y_pred : np.ndarray
        Predicted manifold points (shape: N, ...).
    exp_barycenter_fun : callable
        Function that computes the Exponential Barycenter (Fréchet Mean).

    Returns
    -------
    float
        The Relative Mean Absolute Error (rMAE_mu).
    """
    y_true, y_pred = np.concatenate(y_true, axis=0), np.concatenate(y_pred, axis=0)

    # 1. Compute Frechet Mean (mu) - The optimal trivial predictor
    mu = exp_barycenter_fun(M, y_true)

    # 2. Compute MAE (Numerator)
    # The default jax.vmap(M.metric.dist) applied to two arrays of shape (N, ...) is correct
    vmap_dist = jax.vmap(M.metric.dist)
    mae_vec = vmap_dist(y_true, y_pred)
    MAE = np.mean(mae_vec)

    # --- 3. Compute AvgDev (Denominator) - CORRECTED LINE ---
    # AvgDev = Average distance between True and Mean (the baseline error)
    # Use jax.vmap with in_axes=(None, 0):
    # 'None' broadcasts mu (static point) across all mapped elements of A_true (axis 0).
    dist_mu_vec = jax.vmap(M.metric.dist, in_axes=(None, 0))(mu, y_true)
    AvgDev = np.mean(dist_mu_vec)

    # --- 4. Compute rMAE_mu ---
    if AvgDev < 1e-10:
        return float(0.0)

    rMAE_mu = MAE / AvgDev

    return float(rMAE_mu)

def comp_err(Ytrue, Ypred, n_learn):
    Ytrue = [y[n_learn:] for y in Ytrue]
    rMAE = compute_rmae_mu(M, Ytrue, Ypred, ExponentialBarycenter().compute)
    Ytrue, Ypred = np.concatenate(Ytrue, axis=0), np.concatenate(Ypred, axis=0)
    dim, n = Ytrue.shape[-1], len(Ytrue)
    err_spd, err_euc, rMAE = [None]*n, [None]*n, [None]*n
    id_M = np.eye(dim)
    for k in range(n):
        ytrue, ypred  = Ytrue[k], Ypred[k]
        err_spd += [np.mean([dist(ypred, ytrue)/dist(id_M, ytrue)])]
        err_euc += [np.mean([lg.norm(ypred - ytrue) / lg.norm(ytrue)])]  # log-euclidean
    #mae = [err(Y_pred[k], Ytest[k][n_learn:]).mae() for k in range(len(Ytest))]
    #rMAE = np.array(rMAE)
    print('RMAE_SPD: {:.4f}'.format(np.mean(err_spd)))
    print('RMAE_Euc: {:.4f}'.format(np.mean(err_euc)))
    print('rMAE: {:.4f}'.format(rMAE))
```

```python
# Read and Split Data
Y = generate_rand_trj(n_subj, n_points)
deg = 3
B, Y_fit = fit_poly_dc(M, Y, deg=deg)
#save(Y, B)
#Y, B = load()
#Y = Y_fit
n_train = 16
Y_train, B_train, Y_test = Y[:n_train], B[:n_train], Y[n_train:]
```

```python
n_cp = deg + 1
P = PowerManifold(M, n_cp)
n_learn, n_pred = n_cp, 1
ensemble = VelocityEnsemble(alpha=0.5)
horizon_buffer = 120
pred_args = {
    'n_learn': n_learn,
    'n_pred': n_pred,
    'iterative': False,
    'horizon_buffer': horizon_buffer
}
Ytest = [Y_test[0]]
```

```python
model = Reg(M, lag=True, degree=deg)
Y_pred, metrics = pred(Ytest, model, **pred_args, ensemble_strategy=None)
```

```python
model = Reg(M, lag=True, degree=deg)
ensemble = VelocityEnsemble(alpha=-0.25)
Y_pred, metrics = pred(Ytest, model, **pred_args, ensemble_strategy=ensemble)
```

```python
# Ridge Regression
# Covariance matrix and mean
Ex = ExponentialBarycenter()
mean_b_train = Ex.compute(P, B_train, max_iter=30)
w = jax.vmap(jax.jit(P.metric.log), (None, 0))(mean_b_train, B_train)
# Covariance matrix
cov_b_train = cov_mat(P.metric.log, B_train, mean_b_train)
mu = 1e-4
model = RidgeReg(M, mean_b_train, cov_b_train, mu, lag=True, degree=deg)
Y_pred, metrics = pred(Ytest, model, iterative=False, n_learn=n_learn, n_pred=n_pred, horizon_buffer=horizon_buffer)
#comp_err(Ytest, Y_pred, n_learn)
```

```python
# AutoReg with ARMA
model = ARMA(M, lag=True, order=(2, 0), local=True)
Y_pred, metrics = pred(Ytest, model, **pred_args)
```

```python
# Weighted Average
model = WeightedAverage(M, lag=False, order=2)
Y_pred, metrics = pred(Ytest, model, **pred_args, ensemble_strategy=None)
```
